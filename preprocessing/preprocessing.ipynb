Import Libarary

```python
import cv2
import os
import matplotlib.pyplot as plt
```

 PATH DATASET

```python
dataset_path = "../Dataset Citra"
output_path = "../preprocessed_dataset"

# membuat folder output jika belum ada
os.makedirs(output_path, exist_ok=True)
```

Prepocessing

```python
# =========================
# LOOP SEMUA FOLDER
# =========================
folders = os.listdir(dataset_path)

for folder in folders:

    folder_path = os.path.join(dataset_path, folder)

    # skip kalau bukan folder
    if not os.path.isdir(folder_path):
        continue

    # membuat folder output sesuai kategori
    output_folder = os.path.join(output_path, folder)
    os.makedirs(output_folder, exist_ok=True)

    # ambil semua gambar
    images = os.listdir(folder_path)

    for image_name in images:

        image_path = os.path.join(folder_path, image_name)

        # baca gambar
        img = cv2.imread(image_path)

        # skip jika gambar gagal dibaca
        if img is None:
            print(f"Gagal membaca {image_name}")
            continue

        # =========================
        # RESIZE PROPORSIONAL
        # =========================
        height, width = img.shape[:2]

        new_width = 400
        new_height = int((new_width / width) * height)

        resize = cv2.resize(img, (new_width, new_height))

        # =========================
        # GRAYSCALE
        # =========================
        gray = cv2.cvtColor(resize, cv2.COLOR_BGR2GRAY)

        # =========================
        # GAUSSIAN BLUR
        # =========================
        blur = cv2.GaussianBlur(gray, (5,5), 0)

        # =========================
        # SIMPAN HASIL
        # =========================
        save_path = os.path.join(output_folder, image_name)

        cv2.imwrite(save_path, blur)

        print(f"{image_name} selesai diproses")

print("SEMUA GAMBAR BERHASIL DIPREPROCESSING")
```
